# Data Cleaning and Exploration 

```python
import pandas as pd 
import numpy as np

df_response = pd.read_csv('/Users/shiyuzhang/Desktop/DA_Assignment/response_answers.txt',sep = '|')
df_users = pd.read_csv('/Users/shiyuzhang/Desktop/DA_Assignment/users.txt',sep = '|')
df_places =  pd.read_csv('/Users/shiyuzhang/Desktop/DA_Assignment/places.txt',sep = '|')
df_questions =  pd.read_csv('/Users/shiyuzhang/Desktop/DA_Assignment/questions.txt',sep = '|')
```

## Data Cleaning for response.txt file

```python
# Get a general sense of what the data looks like:
df_response.head()
```

```python
# First check data types of each column and see if there is any wrong data type
df_response.dtypes
```

```python
df_response['question_identity']=df_response['question_identity'].replace('\,','',regex = True)
df_response['question_identity'] = pd.to_numeric(df_response['question_identity'])
```

```python
# Check for missing value for each column
df_response.isnull().sum()

# Based on the result, most of the columns only have one missing value, and response_answer has 23 missing,
# because this is a survey data, I think our best option is to drop these null values, it won't have huge impact on the survey result
```

```python
df_response.dropna(axis = 0)
```

```python
# Based on the unique nature of this dataset, only longitude, latitude and reward points are considered to be numeric value, there
# is no need to check for outliers since it won't make any sense. 
```

```python
df_response.duplicated(subset=None, keep='first').sum()
# There is no duplicates in this dataset
```
